The following function called `plot_decision_boundary` encapsulates the code that we learned in the last part of Tutorial 3. It tests a well-trained ANN `model` with all points in a 2D space. It then calculates the __decision boundary__ and plots it out.

It is called `plot_decision_boundary` with the signature:

```python
def plot_decision_boundary(model, X):
    ....
```

````python
def plot_decision_boundary(model, X):
    hticks = np.linspace(X.min()-0.1, X.max()+0.1, 101)
    vticks = np.linspace(X.min()-0.1, X.max()+0.1, 101)
    aa, bb = np.meshgrid(hticks, vticks)
    ab = np.c_[aa.ravel(), bb.ravel()]
    
    c = model.predict(ab)
    cc = c.reshape(aa.shape)
    
    plt.figure(figsize=(7, 7))
    plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)

    plt.title("Blue circles and Red crosses");
````

````python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000,
                    centers=2,
                    random_state=0)

plt.figure(figsize=(7, 7))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)

plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses");
````

- re-train your model (that was used to learn XOR) on this dataset
- if needed, alter the architecture of the model and/or the number of epochs 
- display the decision boundary for this model by calling the `plot_decision_boundary` above (the following is a sample)

![decision_boundary_blobs.png](./assets/decision_boundary_blobs.png)

````python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


training_data = X
target_data = y

model = Sequential()

model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

model.fit(training_data, target_data, epochs=100, verbose=2, batch_size=4)
````

````python
plot_decision_boundary(model, X)


plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)

plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses");
````
